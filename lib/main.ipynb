```python
fileloc="../data/philosophy_data.csv"
dfdata=pd.read_csv(fileloc)
print(dfdata.loc[0])
```

```python
def tokenizedtxttolst(tokenizedstr):
    wordlst=[]
    tokenizedlen=len(tokenizedstr)
    a=0
    l=0
    r=0
    intxt=False
    while a<tokenizedlen:
        if tokenizedstr[a]=="'":
            if intxt is False:
                l=a
                intxt=True
            else:
                r=a
                intxt=False
                word=tokenizedstr[l+1:r]
                if word not in stopwords.words('english'):
                    wordlst.append(word)
        a+=1
    return wordlst
```

```python
def arraytostr(arr):
    return "['"+"', '".join(arr)+"']"
```

```python
schools=[]
for i in range(len(dfdata)):
    if dfdata.iat[i,2] not in schools:
        schools.append(dfdata.iat[i,2])
print(schools)
```

```python
datadict={}
for school in schools:
    datadict[school]=[]
    filename="../output/"+school+".csv"
    header=["school","processedlst"]
    datalst=[]
    for i in range(len(dfdata)):
        if i%1000==0:
            print(i)
        if dfdata.iat[i,2]==school:
            tmplst=tokenizedtxttolst(dfdata.iat[i,9])
            if len(tmplst)>=3:
                datadict[school].append(tmplst)
    #            datalst.append([school,arraytostr(tmplst)])
    #with open(filename,'w',newline="",encoding="utf-8") as csvfile:
    #    csvwriter = csv.writer(csvfile)
    #    csvwriter.writerow(header)
    #    csvwriter.writerows(datalst)
print(datadict)
```

```python
for school in schools:
    print(datadict[school][0:10])
```

```python
smalldatadict={}
for school in schools:
    smalldatadict[school]=random.sample(datadict[school],1000)
```

```python
corrarr=[]
for i in range(len(schools)):
    temparr=[]
    for j in range(len(schools)):
        temparr.append(0)
    corrarr.append(temparr)
for i in range(len(schools)):
    print(i)
    for j in range(i+1,len(schools)):
        print(j)
        thiscorrarr=[]
        l1=smalldatadict[schools[i]]
        l2=smalldatadict[schools[j]]
        for k in range(1000):
            for t in range(1000):
                thiscorrarr.append(wv.n_similarity(l1[k],l2[t]))
        correlation=np.mean(thiscorrarr)
        corrarr[i][j]=correlation
        corrarr[j][i]=correlation
    print()
print(corrarr)
```

```python
for i in range(len(corrarr)):
    corrarr[i][i]=1
```

```python
dfcorr=pd.DataFrame(corrarr,columns=schools,index=schools)
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(dfcorr, annot=True,ax=ax)
plt.savefig('../figs/results.png')
```
